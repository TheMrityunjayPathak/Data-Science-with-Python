<font color='#6a66bd' size="5px">Introduction to Matplotlib</font>
<br>
<br>
<strong>What is Matplotlib?</strong>
<br>
• Matplotlib is a low level graph plotting library in python that serves as a visualization utility.
<br>
• Matplotlib was created by John D. Hunter.
<br>
• Matplotlib is open source and we can use it freely.
<br>
• Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.
<br>
<br>
<strong>Basic Components of a Plot</strong>
<br>
<i class="fas fa-arrow-right"></i> <strong>Figure :</strong> The overall window or page where you plot everything.
<br>
<i class="fas fa-arrow-right"></i> <strong>Axes :</strong> A single plot or graph inside the figure can have multiple axes in a figure.
<br>
<i class="fas fa-arrow-right"></i> <strong>Axis :</strong> The x-axis and y-axis of the plot.
<br>
<i class="fas fa-arrow-right"></i> <strong>Artist :</strong> Any object that can be drawn such as lines, text and markers.
</div>

<font color='#6a66bd' size="5px">Matplotlib Magic Command</font>
<br>
<br>
<strong>What is %matplotlib inline?</strong>
<br>
• The <code>%matplotlib inline</code> magic command in Jupyter Notebooks is used to configure how plots are displayed.
<br>
• This command ensures that all Matplotlib plots are displayed directly within the Jupyter Notebook’s output cells. 
<br>
• When you use this command the plots are embedded in the notebook itself, rather than appearing in a separate window.
<br>
<br>
<strong>Use of this Command</strong>
<br>
• With <code>%matplotlib inline</code>, you don’t need to explicitly call <code>plt.show()</code> for plots to appear in the notebook.
<br>
• Using <code>plt.show()</code> is still good practice if you want to ensure plots are rendered at any specific cell or if you have multiple plots in a single cell.
<br>
<br>
<strong>How to use the Command</strong>
<br>
• The <code>%matplotlib inline</code> magic command only needs to be executed once per notebook session.
<br>
• It sets up the environment for displaying plots inline for the entire session.
</div>

```python
# Matplotlib Magic Command
%matplotlib inline
```

<font color='#6a66bd' size="5px">Matplotlib Plotting</font>
<br>
<br>
<strong>Plotting X and Y Points</strong>
<br>
• The <code>plot()</code> function is used to draw points in a diagram.
<br>
• By default, the function draws a line from point to point.
<br>
• The function takes parameters for specifying points in the diagram.
<br>
• Parameter 1 is an array containing the points on the x-axis.
<br>
• Parameter 2 is an array containing the points on the y-axis.
<div style="background-color:#ADD8E6; padding:8px; border:1px solid #87CEEB; border-radius:4px;">
<strong>Note :</strong> The x-axis is the horizontal axis.
</div>
<div style="background-color:#ADD8E6; padding:8px; border:1px solid #87CEEB; border-radius:4px;">
<strong>Note :</strong> The y-axis is the vertical axis.
</div>
</div>

```python
# Importing NumPy
import numpy as np

# Defining Data Points
x_points = np.array([2,4,6,8,10,12])
y_points = np.array([12,85,32,7,24,33])

# Plotting Function
plt.plot(x_points, y_points)

# Displaying the Plot
plt.show()
```

<strong>Position the Title</strong>
<br>
• You can use the <code>loc</code> parameter in <code>title()</code> to position the title.
<br>
• Legal values are : 'left', 'right' and 'center'. Default value is 'center'.
</div>

```python
# Defining Data Points
x_points = np.array([2,4,6,8,10,12])
y_points = np.array([12,85,32,7,24,33])

# Plotting Function
plt.plot(x_points, y_points)

# Setting the Labels for X and Y Axis
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

# Setting the Title of the Plot
plt.title("Matplotlib : Line Plot", loc="right")

# Displaying the Plot
plt.show()
```

<div style="background-color:gainsboro; padding:8px; border:2px dotted black; border-radius:8px; font-family:verdana; line-height: 1.7em">
<strong>Set Font Properties for Title and Labels</strong>
<br>
• You can use the <code>fontdict</code> parameter in xlabel(), ylabel() and title() to set font properties for the title and labels.
</div>

```python
# Defining Data Points
x_points = np.array([2,4,6,8,10,12])
y_points = np.array([12,85,32,7,24,33])

# Plotting Function
plt.plot(x_points, y_points)

# Font Properties for the xlabel, ylabel and title
font1 = {'family':'verdana','color':'blue','size':20}
font2 = {'family':'verdana','color':'darkred','size':15}

# Setting the Labels for X and Y Axis
plt.xlabel("X-Axis", fontdict=font2)
plt.ylabel("Y-Axis", fontdict=font2)

# Setting the Title of the Plot
plt.title("Matplotlib : Line Plot", fontdict=font1)

# Displaying the Plot
plt.show()
```

<font color='#6a66bd' size="5px">Figure Size</font>
<br>
<br>
• In Matplotlib, <code>figsize</code> is a parameter used to specify the dimensions of a figure in Matplotlib.
<br>
• It is given as a tuple of two numbers which represent the width and height of the figure.
<br>
<br>
<strong>Figsize Parameter</strong>
<br>
• When creating a new figure, you can specify the size using the <code>figsize</code> parameter in the <code>plt.figure()</code> function.
</div>

```python
# Create a figure with a specific size
fig, ax = plt.subplots(figsize=(12, 6))

# Plotting Points
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plotting Function
ax.plot(x, y)

# Setting the Title of the Plot
ax.set_title('Example Plot with figsize=(12, 6)')

# Setting the Labels for X and Y Axis
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')

# Displaying the Plot
plt.show()
```

<strong>Explode</strong>
<br>
• Explode parameter if specified, makes one of the wedges to stand out.
<br>
• The explode parameter must be an array with one value for each wedge.
<br>
• Each value represents how far from the center each wedge is displayed.

```python
# Plotting Points
x = np.array([35,25,25,15])

# Labels of the Plot
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

# Explode Value for Each Wedge
myexplode = [0.2,0,0,0]

# Plotting Function
plt.pie(x, labels=mylabels, explode=myexplode)

# Displaying the Plot
plt.show() 
```

<strong>Colormap</strong>
<br>
• You can change the colormap using the <code>cmap</code> parameter. 
<br>
• Some common colormaps include <code>'viridis'</code>, <code>'plasma'</code>, <code>'inferno'</code>, <code>'cividis'</code>, <code>'coolwarm'</code>, and <code>'magma'</code>.
</div>

```python
# Plotting Points
data = np.random.rand(10, 10)

# Create a figure with a specific size
fig, ax = plt.subplots()

# Plotting Function
hm = ax.imshow(data, cmap="magma")

# Colorbar for the Plot
cbar = fig.colorbar(hm)

# Displaying the Plot
plt.show()
```

<div style="background-color:gainsboro; padding:8px; border:2px dotted black; border-radius:8px; font-family:verdana; line-height: 1.7em">
<strong>Colorbar</strong>
<br>
• You can customize the colorbar by adjusting its label and ticks.
</div>

```python
# Plotting Points
data = np.random.rand(10, 10)

# Create a figure with a specific size
fig, ax = plt.subplots()

# Plotting Function
hm = ax.imshow(data, cmap="magma")

# Colorbar for the Plot
cbar = fig.colorbar(hm)

# Label for the Colorbar
cbar.set_label('Intensity')

# Label for the Colorbar Ticks
cbar.set_ticks([0, 0.5, 1])

# Displaying the Plot
plt.show()
```

<div style="background-color:gainsboro; padding:8px; border:2px dotted black; border-radius:8px; font-family:verdana; line-height: 1.7em">
<strong>Axis Labels and Title</strong>
<br>
• Adding labels to the Plot and X/Y Axis.
</div>

```python
# Plotting Points
data = np.random.rand(10, 10)

# Create a figure with a specific size
fig, ax = plt.subplots()

# Plotting Function
hm = ax.imshow(data, cmap="magma")

# Colorbar for the Plot
cbar = fig.colorbar(hm)

# Label for the Colorbar
cbar.set_label('Intensity')

# Label for the Colorbar Ticks
cbar.set_ticks([0, 0.5, 1])

# Label for the Plot
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')

# Title for the Plot
ax.set_title('Heatmap Plot')

# Displaying the Plot
plt.show()
```
