<font color='#6a66bd' size="5px">Python Comments</font>
<br>
<br>
<strong>What is Comments?</strong>
<br>
• Comments can be used to explain Python code.
<br>
• Comments can be used to make the code more readable.
<br>
• Comments can be used to prevent execution when testing code.
<br>
<br>
<strong>Creating a Comment</strong>
<br>
• Comments starts with # and Python will ignore them.
<pre><strong>python</strong><code class="language-python">
#this is a comment
</code></pre>
• Comments can be placed at the end of a line and Python will ignore the rest of the line.
<pre><strong>python</strong><code class="language-python">
print("Hello, World!") #this is a comment
</code></pre>
• A comment does not have to be text that explains the code, it can also be used to prevent Python from executing code.
<pre><strong>python</strong><code class="language-python">
#print("Hello, World!")
</code></pre>
<br>
<strong>Multiline Comments</strong>
<br>
• Python does not really have a syntax for multiline comments.
<br>
• To add a multiline comment you could insert a # for each line.
<pre><strong>python</strong><code class="language-python">
#this is a comment
#written in
#more than just one line
</code></pre>
• You can also use a multiline string (docstring).
<pre><strong>python</strong><code class="language-python">
"""
This is a comment
written in
more than just one line
"""
</code></pre>

<font color='#6a66bd' size="5px">Python Variables</font>
<br>
<br>
<strong>What is Variables?</strong>
<br>
• Variables are containers for storing data values.
<br>
<br>
<strong>Creating a Variable</strong>
<br>
• Python has no command for declaring a variable.
<br>
• A variable is created the moment you first assign a value to it.
<pre><strong>python</strong><code class="language-python">
x = 5
y = "John"
print(x)
print(y)
</code></pre>
• Variables do not need to be declared with any particular type and can even change type after they have been set.
<pre><strong>python</strong><code class="language-python">
x = 4 #x is of type int
x = "Sally" #x is now of type str
print(x)
</code></pre>
<br>
<strong>Casting</strong>
<br>
• If you want to specify the data type of a variable, this can be done with casting.
<pre><strong>python</strong><code class="language-python">
x = str(3) #x will be '3'
y = int(3) #y will be 3
z = float(3) #z will be 3.0
</code></pre>
<br>
<strong>Get the Type</strong>
<br>
• You can get the data type of a variable with the type() function.
<pre><strong>python</strong><code class="language-python">
x = 5
y = "John"
print(type(x))
print(type(y))
</code></pre>
<br>
<strong>Case Sensitive</strong>
<br>
• Variable names are case-sensitive.
<pre><strong>python</strong><code class="language-python">
a = 4
A = "Sally"
#A will not overwrite a
</code></pre>
<br>
<strong>Naming Python Variable</strong>
<br>
• A variable name must start with a letter or the underscore character.
<br>
• A variable name cannot start with a number.
<br>
• A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ ).
<br>
• Variable names are case-sensitive (age, Age and AGE are three different variables).
<br>
• A variable name cannot be any of the Python keywords.
<br>
<strong>Legal Variable Names</strong>
<div style="background-color:lightgreen; padding:8px; border:1px solid green; border-radius:4px;">
myvar = "John"
<br>
my_var = "John"
<br>
_my_var = "John"
<br>
myVar = "John"
<br>
MYVAR = "John"
<br>
myvar2 = "John"
</div>
<strong>IlLegal Variable Names</strong>
<div style="background-color:#F88379; padding:8px; border:1px solid red; border-radius:4px;">
2myvar = "John"
<br>
my-var = "John"
<br>
my var = "John"
</div>
<br>
<strong>Multi Words Variable Names</strong>
<br>
• Variable names with more than one word can be difficult to read.
<br>
• There are several techniques you can use to make them more readable.
<br>
<strong>Camel Case</strong>
<br>
• Each word except the first starts with a capital letter.
<pre><strong>python</strong><code class="language-python">
myVariableName = "John"
</code></pre>
<strong>Pascal Case</strong>
<br>
• Each word starts with a capital letter.
<pre><strong>python</strong><code class="language-python">
MyVariableName = "John"
</code></pre>
<strong>Snake Case</strong>
<br>
• Each word is separated by an underscore character.
<pre><strong>python</strong><code class="language-python">
my_variable_name = "John"
</code></pre>
<br>
<strong>Assign Multiple Values</strong>
<br>
• Python allows you to assign values to multiple variables in one line.
<pre><strong>python</strong><code class="language-python">
x, y, z = "Orange", "Banana", "Cherry"
print(x)
print(y)
print(z)
</code></pre>
• You can assign the same value to multiple variables in one line also.
<pre><strong>python</strong><code class="language-python">
x = y = z = "Orange"
print(x)
print(y)
print(z)
</code></pre>
<div style="background-color:#ADD8E6; padding:8px; border:1px solid #87CEEB; border-radius:4px;">
<strong>Note :</strong> Make sure the number of variables matches the number of values or else you will get an error.
</div>
<br>
<strong>Output Variables</strong>
<br>
• The Python print() function is often used to output variables.
<pre><strong>python</strong><code class="language-python">
x = "Python is awesome"
print(x)
</code></pre>
• In the print() function, you output multiple variables separated by a comma.
<pre><strong>python</strong><code class="language-python">
x = "Python"
y = "is"
z = "awesome"
print(x, y, z)
</code></pre>
• You can also use the + operator to output multiple variables.
<pre><strong>python</strong><code class="language-python">
x = "Python "
y = "is "
z = "awesome"
print(x + y + z)
</code></pre>
• For numbers, the + character works as a mathematical operator.
<pre><strong>python</strong><code class="language-python">
x = 5
y = 10
print(x + y)
</code></pre>
• The best way to output multiple variables in the print() function is to separate them with commas which even support different data types.
<pre><strong>python</strong><code class="language-python">
x = 5
y = "John"
print(x, y)
</code></pre>
<br>
<strong>Global Variables</strong>
<br>
• Variables that are created outside of a function are known as global variables.
<br>
• Global variables can be used everywhere both inside of function and outside.
<pre><strong>python</strong><code class="language-python">
x = "awesome"
def myfunc():
  print("Python is " + x)
myfunc()
</code></pre>
• If you create a variable with the same name inside a function, this variable will be local and can only be used inside the function.
<br>
• The global variable with the same name will remain as it was and with the original value.
<pre><strong>python</strong><code class="language-python">
x = "awesome"
def myfunc():
  x = "fantastic"
  print("Python is " + x)
myfunc()
print("Python is " + x)
</code></pre>
• Normally, when you create a variable inside a function that variable is local and can only be used inside that function.
<br>
• To create a global variable inside a function you can use the global keyword.
<pre><strong>python</strong><code class="language-python">
def myfunc():
  global x
  x = "fantastic"
myfunc()
print("Python is " + x)
</code></pre>
• Also, use the global keyword if you want to change a global variable inside a function.
<pre><strong>python</strong><code class="language-python">
x = "awesome"
def myfunc():
  global x
  x = "fantastic"
myfunc()
print("Python is " + x)
</code></pre>

<font color='#6a66bd' size="5px">Python Numbers</font>
<br>
<br>
• There are three numeric types in Python that are int, float and complex.
<br>
• Variables of numeric types are created when you assign a value to them.
<pre><strong>python</strong><code class="language-python">
x = 1 #int
y = 2.8 #float
z = 1j #complex
</code></pre>
• To verify the type of any object in Python use the type() function.
<pre><strong>python</strong><code class="language-python">
print(type(x))
print(type(y))
print(type(z))
</code></pre>
• Int or Integer is a whole number positive or negative without decimals of unlimited length.
<pre><strong>python</strong><code class="language-python">
x = 1
y = 35656222554887711
z = -3255522
</code></pre>
• Float or Floating Point Number is a number, positive or negative containing one or more decimals.
<pre><strong>python</strong><code class="language-python">
x = 1.10
y = 1.0
z = -35.59
</code></pre>
• Complex numbers are written with a "j" as the imaginary part.
<pre><strong>python</strong><code class="language-python">
x = 3+5j
y = 5j
z = -5j
</code></pre>
<br>
<strong>Type Conversion</strong>
<br>
• You can convert numbers from one type to another with the int(), float() and complex() methods.
<pre><strong>python</strong><code class="language-python">
x = 1 #int
y = 2.8 #float
z = 1j #complex
#convert from int to float
a = float(x)
#convert from float to int
b = int(y)
#convert from int to complex
c = complex(x)
</code></pre>
<div style="background-color:#ADD8E6; padding:8px; border:1px solid #87CEEB; border-radius:4px;">
<strong>Note :</strong> You cannot convert complex numbers into another number type.
</div>
<br>
<strong>Random Number</strong>
<br>
• Python does not have a random() function to make a random number but Python has a built-in module called random that can be used to make random numbers.
<br>
<br>
<strong>List of Avaliable Methods for Random Numbers</strong>
<br>
<br>
<strong>random() Method</strong>
<br>
• The random() method returns a random floating number between 0 and 1.
<pre><strong>python</strong><code class="language-python">
import random
print(random.random())
</code></pre>
<br>
<strong>randrange() Method</strong>
<br>
• The randrange() method returns a randomly selected element from the specified range.
<br>
<strong>Parameter Values</strong>
<br>
• start	(Optional) - An integer specifying at which position to start. Default 0.
<br>
• stop (Required) - An integer specifying at which position to end.
<br>
• step (Optional) - An integer specifying the incrementation. Default 1.
<pre><strong>python</strong><code class="language-python">
import random
print(random.randrange(3, 9))
</code></pre>
<br>
<strong>randint() Method</strong>
<br>
• The randint() method returns an integer number selected element from the specified range.
<br>
<strong>Parameter Values</strong>
<br>
• start	(Required) - An integer specifying at which position to start.
<br>
• stop (Required) - An integer specifying at which position to end.
<pre><strong>python</strong><code class="language-python">
import random
print(random.randint(3, 9))
</code></pre>
<br>
<strong>choice() Method</strong>
<br>
• The choice() method returns a randomly selected element from the specified sequence.
<br>
• The sequence can be a string, a range, a list, a tuple or any other kind of sequence.
<pre><strong>python</strong><code class="language-python">
import random
mylist = ["apple", "banana", "cherry"]
print(random.choice(mylist))
</code></pre>
<br>
<strong>shuffle() Method</strong>
<br>
• The shuffle() method takes a sequence like a list and reorganize the order of the items.
<pre><strong>python</strong><code class="language-python">
import random
mylist = ["apple", "banana", "cherry"]
random.shuffle(mylist)
print(mylist)
</code></pre>
<div style="background-color:#ADD8E6; padding:8px; border:1px solid #87CEEB; border-radius:4px;">
<strong>Note :</strong> This method changes the original list it does not return a new list.
</div>
<br>
<strong>uniform() Method</strong>
<br>
• The uniform() method returns a random floating number between the two specified numbers (both included).
<pre><strong>python</strong><code class="language-python">
import random
print(random.uniform(20, 60))
</code></pre>
</div>

<font color='#6a66bd' size="5px">Python Casting</font>
<br>
<br>
<strong>Specify a Variable Type</strong>
<br>
• There may be times when you want to specify a type on to a variable. This can be done with casting.
<br>
• Python is an object-orientated language and as such it uses classes to define data types including its primitive types.
<br>
• Casting in python is therefore done using constructor functions:
<br>
<br>
<strong>Integer</strong> 
<br>
• Constructs an integer number from an integer literal, a float literal (by removing all decimals) or a string literal (providing the string represents a whole number).
<pre><strong>python</strong><code class="language-python">
x = int(1) #x will be 1
y = int(2.8) #y will be 2
z = int("3") #z will be 3
</code></pre>
<strong>Float</strong>
<br>
• Constructs a float number from an integer literal, a float literal or a string literal (providing the string represents a float or an integer).
<pre><strong>python</strong><code class="language-python">
x = float(1) #x will be 1.0
y = float(2.8) #y will be 2.8
z = float("3") #z will be 3.0
w = float("4.2") #w will be 4.2
</code></pre>
<strong>String</strong>
<br>
• Constructs a string from a wide variety of data types including strings, integer literals and float literals.
<pre><strong>python</strong><code class="language-python">
x = str("s1") #x will be 's1'
y = str(2) #y will be '2'
z = str(3.0) #z will be '3.0'
</code></pre>

<font color='#6a66bd' size="5px">Python Strings</font>
<br>
<br>
<strong>Strings</strong>
<br>
• Strings in python are surrounded by either single quotation marks or double quotation marks.
<br>
• 'hello' is the same as "hello".
<br>
• You can display a string literal with the print() function.
<pre><strong>python</strong><code class="language-python">
print("Hello")
print('Hello')
</code></pre>
<br>
<strong>Quotes Inside Quotes</strong>
<br>
• You can use quotes inside a string as long as they don't match the quotes surrounding the string.
<pre><strong>python</strong><code class="language-python">
print("It's alright")
print("He is called 'Johnny'")
print('He is called "Johnny"')
</code></pre>
<br>
<strong>Assign String to a Variable</strong>
<br>
• Assigning a string to a variable is done with the variable name followed by an equal sign and the string.
<pre><strong>python</strong><code class="language-python">
a = "Hello"
print(a)
</code></pre>
<br>
<strong>Multiline Strings</strong>
<br>
• You can assign a multiline string to a variable by using three quotes.
<pre><strong>python</strong><code class="language-python">
#you can use three double quotes
a = """Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua."""
print(a)
</code></pre>
<pre><strong>python</strong><code class="language-python">
#or three single quotes
a = '''Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua.'''
print(a)
</code></pre>
<div style="background-color:#ADD8E6; padding:8px; border:1px solid #87CEEB; border-radius:4px;">
<strong>Note :</strong> In the result the line breaks are inserted at the same position as in the code.
</div>
<br>
<strong>Strings are Arrays</strong>
<br>
• Like many other popular programming languages strings in Python are arrays of bytes representing unicode characters.
<br>
• However Python does not have a character data type a single character is simply a string with a length of 1.
<br>
• Square brackets can be used to access elements of the string.
<pre><strong>python</strong><code class="language-python">
a = "Hello, World!"
print(a[1])
</code></pre>
<div style="background-color:#ADD8E6; padding:8px; border:1px solid #87CEEB; border-radius:4px;">
<strong>Note :</strong> Remember that the first character has the position 0.
</div>
<br>
<strong>Looping Through a String</strong>
<br>
• Since strings are arrays, we can loop through the characters in a string with a for loop.
<pre><strong>python</strong><code class="language-python">
for x in "banana":
  print(x)
</code></pre>
<br>
<strong>String Length</strong>
<br>
• To get the length of a string use the len() function.
<pre><strong>python</strong><code class="language-python">
a = "Hello, World!"
print(len(a))
</code></pre>
<br>
<strong>Check String</strong>
<br>
• To check if a certain phrase or character is present in a string, we can use the in keyword.
<pre><strong>python</strong><code class="language-python">
txt = "The best things in life are free!"
print("free" in txt)
</code></pre>
• To check if a certain phrase or character is NOT present in a string, we can use the not in keyword.
<pre><strong>python</strong><code class="language-python">
txt = "The best things in life are free!"
print("expensive" not in txt)
</code></pre>
<br>
<strong>String Slicing</strong>
<br>
• You can return a range of characters by using the slice syntax.
<br>
• Specify the start index and the end index separated by a colon to return a part of the string.
<pre><strong>python</strong><code class="language-python">
b = "Hello, World!"
print(b[2:5])
</code></pre>
<div style="background-color:#ADD8E6; padding:8px; border:1px solid #87CEEB; border-radius:4px;">
<strong>Note :</strong> The first character has index 0.
</div>
<br>
<strong>Slice From the Start</strong>
<br>
• By leaving out the start index the range will start at the first character.
<pre><strong>python</strong><code class="language-python">
b = "Hello, World!"
print(b[:5])
</code></pre>
<br>
<strong>Slice To the End</strong>
<br>
• By leaving out the end index the range will go to the end.
<pre><strong>python</strong><code class="language-python">
b = "Hello, World!"
print(b[2:])
</code></pre>
<br>
<strong>Negative Indexing</strong>
<br>
• Use negative indexes to start the slice from the end of the string.
<pre><strong>python</strong><code class="language-python">

<strong>What are Booleans?</strong>

• In programming you often need to know if an expression is True or False.
<br>
• You can evaluate any expression in Python and get one of two answers True or False.
<br>
• When you compare two values the expression is evaluated and Python returns the Boolean answer.

<br>

<br>
• Any number is True, except 0.
<br>
• Any list, tuple, set and dictionary are True, except empty ones.
<br>
<br>
<strong>Some Values are False</strong>
<br>
• In fact, there are not many values that evaluate to False except empty values such as (), [], {}, "", the number 0, and the value None.
<br>
• And of course the value False evaluates to False.

<strong>Multiply Tuples</strong>

<font color='#6a66bd' size="5px">Python Conditions</font>

• Python supports the usual logical conditions from mathematics.

→ Equals : <code>a == b</code>
<br>
→ Not Equals : <code>a != b</code>
<br>
→ Less than : <code>a &lt; b</code>
<br>
→ Less than or equal to : <code>a &lt;= b</code>
<br>
→ Greater than : <code>a > b</code>
<br>
→ Greater than or equal to : <code>a >= b</code>
<br>
<br>
<strong>If Statement</strong>
<br>
• An "if statement" is written by using the if keyword.
<pre><strong>python</strong><code class="language-python">
a = 33

if b > a:
  print("b is greater than a")
</code></pre>

<br>
• The elif keyword is Python's way of saying "if the previous conditions were not true, then try this condition".
<pre><strong>python</strong><code class="language-python">
a = 33
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")

<br>
• The else keyword catches anything which isn't caught by the preceding conditions.
<pre><strong>python</strong><code class="language-python">
a = 200
b = 33
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
</code></pre>

<strong>Short Hand If Statement</strong>
<br>
• If you have only one statement to execute, you can put it on the same line as the if statement.

if a > b: print("a is greater than b")
</code></pre>

a = 2

print("A") if a > b else print("B")
</code></pre>

• With the while loop we can execute a set of statements as long as a condition is true.
<pre><strong>python</strong><code class="language-python">
i = 1
while i < 6:
  print(i)
  i += 1
</code></pre>

<br>
• With the break statement we can stop the loop even if the while condition is true.
<pre><strong>python</strong><code class="language-python">
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
</code></pre>

• With the continue statement we can stop the current iteration and continue with the next.

  if i == 3:

<font color='#6a66bd' size="5px">Python Functions</font>
<br>
<br>
• A function is a block of code which only runs when it is called.

<br>
• A function can return data as a result.
<br>
<br>
<strong>Creating a Function</strong>
<br>
• In Python a function is defined using the def keyword.
<pre><strong>python</strong><code class="language-python">
def my_function():
  print("Hello from a function")
</code></pre>
<br>
<strong>Calling a Function</strong>
<br>
• To call a function use the function name followed by parenthesis.
<pre><strong>python</strong><code class="language-python">
def my_function():
  print("Hello from a function")
my_function()
</code></pre>
<br>
<strong>Arguments</strong>
<br>
• Information can be passed into functions as arguments.
<br>
• Arguments are specified after the function name inside the parentheses. You can add as many arguments as you want just separate them with a comma.
<pre><strong>python</strong><code class="language-python">
def my_function(fname):
  print(fname + " Refsnes")
my_function("Emil")
my_function("Tobias")
my_function("Linus")
</code></pre>
<div style="background-color:#ADD8E6; padding:8px; border:1px solid #87CEEB; border-radius:4px;">

<br>
• If your function expects 2 arguments you have to call the function with 2 arguments not more and not less.

<br>
• If you do not know how many arguments that will be passed into your function add a * before the parameter name in the function definition.
<br>

<pre><strong>python</strong><code class="language-python">

  print("The youngest child is " + kids[2])
my_function("Emil", "Tobias", "Linus")
</code></pre>

<strong>Keyword Arguments</strong>
<br>
• You can also send arguments with the key = value syntax.
<br>

<font color='#6a66bd' size="5px">Python Scope</font>
<br>
<br>
• A variable is only available from inside the region it is created. This is called scope.
<br>
<br>
<strong>Local Scope</strong>
<br>
• A variable created inside a function belongs to the local scope of that function and can only be used inside that function.
<pre><strong>python</strong><code class="language-python">
def myfunc():
  x = 300
  print(x)
myfunc()
</code></pre>
<br>
<strong>Global Scope</strong>
<br>
• A variable created in the main body of the Python code is a global variable and belongs to the global scope.
<br>
• Global variables are available from within any scope global and local.
<pre><strong>python</strong><code class="language-python">
x = 300
def myfunc():
  print(x)
myfunc()
print(x)
</code></pre>
<br>
<strong>Naming Variable</strong>
<br>
• If you operate with the same variable name inside and outside of a function Python will treat them as two separate variables one available in the global scope (outside the function) and one available in the local scope (inside the function).
<pre><strong>python</strong><code class="language-python">
x = 300
def myfunc():
  x = 200
  print(x)
myfunc()
print(x)
</code></pre>
<br>
<strong>Global Keyword</strong>
<br>
• If you need to create a global variable but are stuck in the local scope you can use the global keyword.
<br>
• The global keyword makes the variable global.
<pre><strong>python</strong><code class="language-python">
def myfunc():
  global x
  x = 300
myfunc()
print(x)
</code></pre>

<font color='#6a66bd' size="5px">Python Try Except</font>
<br>
<br>
• The try block lets you test a block of code for errors.
<br>
• The except block lets you handle the error.
<br>
• The else block lets you execute code when there is no error.
<br>
• The finally block lets you execute code regardless of the result of the try- and except blocks.
<br>
<br>
<strong>Exception Handling</strong>
<br>
• When an error occurs or exception as we call it, Python will normally stop and generate an error message.
<br>
• These exceptions can be handled using the try statement.
<br>
• Without the try block the program will crash and raise an error.
<pre><strong>python</strong><code class="language-python">
try:
  print(x)
except:
  print("An exception occurred")
</code></pre>
<div style="background-color:#ADD8E6; padding:8px; border:1px solid #87CEEB; border-radius:4px;">
<strong>Note :</strong> Since the try block raises an error the except block will be executed.
</div>
<br>
<strong>Many Exceptions</strong>
<br>
• You can define as many exception blocks as you want e.g. if you want to execute a special block of code for a special kind of error.
<br>
• Print one message if the try block raises a NameError and another for other errors.
<pre><strong>python</strong><code class="language-python">
try:
  print(x)
except NameError:
  print("Variable x is not defined")
except:
  print("Something else went wrong")
</code></pre>
<br>
<strong>Else</strong>
<br>
• You can use the else keyword to define a block of code to be executed if no errors were raised.
<pre><strong>python</strong><code class="language-python">
try:
  print("Hello")
except:
  print("Something went wrong")
else:
  print("Nothing went wrong")
</code></pre>
<br>
<strong>Finally</strong>
<br>
• The finally block if specified will be executed regardless if the try block raises an error or not.
<pre><strong>python</strong><code class="language-python">
try:
  print(x)
except:
  print("Something went wrong")
finally:
  print("The 'try except' is finished")
</code></pre>
<br>
<strong>Raise an Exception</strong>
<br>
• As a Python developer you can choose to throw an exception if a condition occurs.
<br>
• To throw (or raise) an exception use the raise keyword.
<pre><strong>python</strong><code class="language-python">
x = -1
if x < 0:
  raise Exception("Sorry, no numbers below zero")
</code></pre>
• You can define what kind of error to raise and the text to print to the user.
<pre><strong>python</strong><code class="language-python">
x = "hello"
if not type(x) is int:
  raise TypeError("Only integers are allowed")
</code></pre>

<font color='#6a66bd' size="5px">Python File Handling</font>
<br>
<br>
• Python has several functions for creating, reading, updating and deleting files.
<br>
• The key function for working with files in Python is the open() function.
<br>
• The open() function takes two parameters : filename and mode.
<br>
• There are four different methods for opening a file :
<br>
→ Read (r) - Default value. Opens a file for reading, error if the file does not exist.
<br>
→ Append (a) - Opens a file for appending, creates the file if it does not exist.
<br>
→ Write (w) - Opens a file for writing, creates the file if it does not exist
<br>
→ Create (x) - Creates the specified file, returns an error if the file exists.
<br>
• In addition you can specify if the file should be handled as binary or text mode :
<br>
→ Text (t) (Default value) - Text mode
<br>
→ Binary (b) - Binary mode
<br>
<br>
<strong>Opening a File</strong>
<br>
• To open a file for reading it is enough to specify the name of the file.
<pre><strong>python</strong><code class="language-python">
f = open("demofile.txt")
</code></pre>
• But you can also do that.
<pre><strong>python</strong><code class="language-python">
f = open("demofile.txt", "rt")
</code></pre>
<div style="background-color:#ADD8E6; padding:8px; border:1px solid #87CEEB; border-radius:4px;">
<strong>Note :</strong> Here "r" for read and "t" for text are the default values, you do not need to specify them.
</div>
<br>
<strong>Opening a File on the Server</strong>
<br>
• Assume we have the following file demofile.txt located in the same folder as Python.
<br>
• To open the file use the built-in open() function.
<br>
• The open() function returns a file object which has a read() method for reading the content of the file.
<pre><strong>python</strong><code class="language-python">
f = open("demofile.txt", "r")
print(f.read())
</code></pre>
• If the file is located in a different location you will have to specify the file path.
<pre><strong>python</strong><code class="language-python">
f = open("D:\\myfiles\demofile.txt", "r")
print(f.read())
</code></pre>
<br>
<strong>Read Only Parts of the File</strong>
<br>
• By default the read() method returns the whole text but you can also specify how many characters you want to return.
<pre><strong>python</strong><code class="language-python">
f = open("demofile.txt", "r")
print(f.read(5))
</code></pre>
<br>
<strong>Read Lines</strong>
<br>
• You can return one line by using the readline() method.
<pre><strong>python</strong><code class="language-python">
f = open("demofile.txt", "r")
print(f.readline())
</code></pre>
• By calling readline() two times you can read the two first lines.
<pre><strong>python</strong><code class="language-python">
f = open("demofile.txt", "r")
print(f.readline())
print(f.readline())
</code></pre>
<br>
<strong>Close Files</strong>
<br>
• It is a good practice to always close the file when you are done with it.
<pre><strong>python</strong><code class="language-python">
f = open("demofile.txt", "r")
print(f.readline())
f.close()
</code></pre>
<br>
<strong>Write a File</strong>
<br>
• To write to an existing file you must add a parameter to the open() function.
<br>
→ Append (a) - will append to the end of the file
<br>
→ Write (w) - will overwrite any existing content
<pre><strong>python</strong><code class="language-python">
f = open("demofile2.txt", "a")
f.write("Now the file has more content!")
f.close()
#open and read the file after appending
f = open("demofile2.txt", "r")
print(f.read())
</code></pre>
<br>
<strong>Create a New File</strong>
<br>
• To create a new file in Python use the open() method with one of the following parameters :
<br>
→ Create (x) - will create a file, returns an error if the file exist.
<br>
→ Append (a) - will create a file, if the specified file does not exist.
<br>
→ Write (w) - will create a file, if the specified file does not exist.
<pre><strong>python</strong><code class="language-python">
f = open("myfile.txt", "x")
</code></pre>
<br>
<strong>Delete a File</strong>
<br>
• To delete a file you must import the OS module and run its os.remove() function.
<pre><strong>python</strong><code class="language-python">
import os
os.remove("demofile.txt")
</code></pre>
• To delete an entire folder, use the os.rmdir() method.
<pre><strong>python</strong><code class="language-python">
import os
os.rmdir("myfolder")
</code></pre>
